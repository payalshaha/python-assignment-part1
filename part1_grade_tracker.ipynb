```python
# Task 1 — Data Parsing & Profile Cleaning
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:
   
   # Clean name (remove trailing whitespace and Title case)
   clean_name = student["name"].strip().title()

   # Convert roll to int
   roll_no = int(student["roll"])

   # Convert marks tsring to list 
   marks_list = [int(i) for i in student["marks_str"].split(", ")]
   
   # Validate name
   words=clean_name.split()
    
   valid_name = True

   for word in words:
       if not word.isalpha():
          valid_name = False

   if valid_name:
     print("✓ Valid name")
   else:
     print("✗ Invalid name")

   cleaned_students.append({
        "name": clean_name,
        "roll": roll_no,
        "marks": marks_list
    })

   print("================================")
   print(f"Student : {clean_name}")
   print(f"Roll No : {roll_no}")
   print(f"Marks   : {marks_list}")
   print("================================")

for s in cleaned_students:
   if s["roll"]== 103:
      
      print("\nStudent with Roll no. 103:")
      print("UPPERCASE: " , s["name"].upper())
      print("lowercase: " , s["name"].lower())
```

```python
# Task 2 — Marks Analysis Using Loops & Conditionals

print("\n=== TASK 2: MARKS ANALYSIS===")

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]


print("\nSubject Grades:")

for i in range(len(subjects)):

    

    # Grade logic
    if marks[i] >= 90:
        grade = "A+"
    elif marks[i] >= 80:
        grade = "A"
    elif marks[i] >= 70:
        grade = "B"
    elif marks[i] >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{subjects[i]} : {marks[i]} -> {grade}")


# Total Marks
total_marks=sum(marks)

# Average Marks
average_marks = total_marks/len(marks)

print("\nTotal Marks: " ,total_marks)
print("Average Marks: " , average_marks)

# Highest Scoring Subject
highest_mark = max(marks)
highest_index = marks.index(highest_mark)

print(f"Highest Subject:  {subjects[highest_index]} --> {lowest_index}")


# Lowest scoring Marks
lowest_mark = min(marks)
lowest_index = marks.index(lowest_mark)

print(f"Lowest Subject:  {subjects[lowest_index]} --> {lowest_mark}")

# ----------------------
# While Loop for new Subjects
# ----------------------

print("\n Add new subjects")
new_subject_count=0

while True:

    subject_name = input("Enter subject name: ")

    if subject_name.lower() == "done":
        break

    marks_input = input("Enter marks 0 to 100:")

    if not marks_input.isdigit():
        print("Invalid Input, MArks must be numeric")
        continue

    marks_value = int(marks_input)

    if(marks_value < 0 or marks_value > 100):
        print("Invalid marks! Must be between 0 and 100")
        continue

    subjects.append(subject_name)
    marks.append(marks_value)
    new_subject_count +=1


print("\nNew Subjects added: ", new_subject_count)
updated_average = sum(marks)/len(marks)
print("Updated average: ", updated_average)
    



```

```python
# Task 3 — Class Performance Summary
print("\n=== TASK 3: CLASS PERFORMANCE SUMMARY ===")

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("\nName               | Average | Status")
print("-------------------------------")

pass_count = 0
fail_count = 0

averages = []
topper_name = ""
topper_average = 0

for name, marks_list in class_data:
    avg= round(sum(marks_list) / len(marks_list),2)
    averages.append(avg)

    if avg>=60:
        status="Pass"
        pass_count +=1
    else:
        status ="Fail"
        fail_count +=1

    print(f"{name:<18} | {avg:^7} | {status}")

    if avg > topper_average:
        topper_average = avg
        topper_name = name


class_average = round(sum(averages) / len(averages),2)

print("\nStudents Passed: ", pass_count)
print("Students Failed: ", fail_count)
print("Class Average: ", class_average)
```

```python
# Task 4 — String Manipulation Utility

print("\nString Operations")

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Step 1- Strip whitespace
clean_essay = essay.strip()
print("\n1. Stripped Essay:")
print(clean_essay)

# Step 2- Title Case
title_case = clean_essay.title()
print("\n2. Title Case:")
print(title_case)

# Step 3-Count word "python"
count_python = clean_essay.count("python")

print("\n3. Count of 'python': ", count_python)

# Step 4-Replace word
replaced_text= clean_essay.replace("python" , "Python")
print("\n4. Replaced text: ", replaced_text)

# Step 5- Split sentences
sentences = clean_essay.split(". ")

print("\n5. Sentence List:" , sentences)

# Step 6- Number Senetnces
print("\n6. Numbered Sentences")
for i in range(len(sentences)):

    sentence = sentences[i]

    if not sentence.endswith("."):
        sentence += "."

    print(f"{i+1}. {sentence}")
```
